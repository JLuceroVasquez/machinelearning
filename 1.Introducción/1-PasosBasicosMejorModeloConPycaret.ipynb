# Paso 1: Cargamos Pycaret y un Dataset

```python
import pycaret
pycaret.__version__
```

```python
from pycaret.datasets import get_data
Datos_traffic = get_data(dataset = 'diabetes') #Aqu√≠ cambias el Dataset
```

# Paso 2: Importar las Regresiones de Pycaret y Crear el Objeto

```python
from pycaret.regression import *
```

```python
# import RegressionExperiment and init the class
from pycaret.regression import RegressionExperiment
ExperimentoRegresion = RegressionExperiment() #Instanciamos la clase
```

# Paso 3: Configurar el Experimento u Objeto

```python
# init setup on exp
ExperimentoRegresion.setup(Datos_traffic, target = 'Class variable', session_id = 123)
                                          #Coloco la etiqueta de mi Target, en este caso el Volumen del Tr√°fico
```

Usamos la VALIDACI√ìN CRUZADA KFold, cuya t√©cnica divide el conjunto de datos en K partes (o folds). En cada iteraci√≥n del proceso, se utiliza uno de los folds como conjunto de prueba y los otros K-1 como conjunto de entrenamiento. Esto se repite K veces.

# Paso 4: Entrenamos y Comparamos los Modelos

```python
#Comparamos los Modelos usando POO
MejorModelo = ExperimentoRegresion.compare_models()
```

# Paso 4 (Opcional):Ordenamiento de los mejores modelos respecto de una m√©trica

```python
Mejores_Modelos_TOP3_MAE = ExperimentoRegresion.compare_models(sort = 'R2', n_select = 3)
```

# Paso 5: Tuneamos los hypermar√°metros del Mejor Modelo

```python
# Ahora seleccionas uno de esos modelos para tunearlo
MejorModelo = Mejores_Modelos_TOP3_MAE[0]  # Selecciona el mejor de los tres
```

```python
MejorModelo_Tuneado = ExperimentoRegresion.tune_model(estimator=MejorModelo)
```
